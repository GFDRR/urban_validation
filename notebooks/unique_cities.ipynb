```python
import pandas as pd
from pathlib import Path
```

```python
csv_path = Path("../raster_all_cities_merged.csv")
df = pd.read_csv(csv_path)
print("Columns:", df.columns.tolist())
```

```python
df["pred_area_total_m2"] = pd.to_numeric(
    # df["pred_area_total_m2"],
    df["tp_m2"],
    errors="coerce"
)
df.head()
```

```python
filtered = df[
    (df["dataset"] == "obt_2023") &
    # (df["pred_area_total_m2"].notna()) &
    (df["tp_m2"].notna()) &
    (df["pred_area_total_m2"] != 0)
]
filtered.head()
```

```python
unique_cities = sorted(filtered["city"].dropna().unique())

print(f"\nNumber of matching rows: {len(filtered)}")
print(f"Number of unique cities: {len(unique_cities)}\n")
```
